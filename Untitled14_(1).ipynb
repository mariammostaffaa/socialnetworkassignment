```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
edges_path = "facebook_combined.txt"

edges = pd.read_csv(edges_path, sep=" ", header=None, names=["src", "dst"])

G = nx.from_pandas_edgelist(edges, source="src", target="dst")

print("Nodes:", G.number_of_nodes())
print("Edges:", G.number_of_edges())
```

```python
import numpy as np
from networkx.algorithms.community import greedy_modularity_communities

def compute_features(G):
    # degree
    degree_dict = dict(G.degree())
    # clustering coefficient
    clustering_dict = nx.clustering(G)
    # centralities
    deg_cent = nx.degree_centrality(G)
    bet_cent = nx.betweenness_centrality(G, k=200, seed=42)
    close_cent = nx.closeness_centrality(G)

    # communities
    communities = list(greedy_modularity_communities(G))
    community_dict = {}
    for cid, comm in enumerate(communities):
        for node in comm:
            community_dict[node] = cid

    nodes = list(G.nodes())
    df = pd.DataFrame({
        "node": nodes,
        "degree": [degree_dict[n] for n in nodes],
        "clustering": [clustering_dict[n] for n in nodes],
        "deg_cent": [deg_cent[n] for n in nodes],
        "bet_cent": [bet_cent[n] for n in nodes],
        "close_cent": [close_cent[n] for n in nodes],
        "community": [community_dict[n] for n in nodes],
    })
    return df

features_df = compute_features(G)
features_df.head()
```

```python
np.random.seed(42)

nodes = features_df["node"].values
num_nodes = len(nodes)
bot_ratio = 0.10   # 10% bots
num_bots = int(bot_ratio * num_nodes)

bot_nodes = np.random.choice(nodes, size=num_bots, replace=False)

# 0 = human, 1 = bot
labels = np.zeros(num_nodes, dtype=int)
labels[np.isin(nodes, bot_nodes)] = 1

features_df["label"] = labels

features_df["label"].value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

feature_cols = ["degree", "clustering", "deg_cent", "bet_cent", "close_cent", "community"]

X = features_df[feature_cols].values
y = features_df["label"].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

clf_base = RandomForestClassifier(n_estimators=200, random_state=42)
clf_base.fit(X_train, y_train)

y_pred = clf_base.predict(X_test)

print("=== Baseline (no attack) ===")
print(classification_report(y_test, y_pred))
```

```python
import random

G_struct = G.copy()

selected_bots = bot_nodes[: len(bot_nodes)//2]

human_nodes = list(features_df[features_df["label"] == 0]["node"])

for b in selected_bots:
    new_neighbors = random.sample(human_nodes, 5)
    for h in new_neighbors:
        if not G_struct.has_edge(b, h):
            G_struct.add_edge(b, h)


    existing = list(G_struct.neighbors(b))
    if len(existing) > 5:
        to_remove = random.sample(existing, 3)
        for r in to_remove:
            if G_struct.has_edge(b, r):
                G_struct.remove_edge(b, r)
```

```python
features_struct = compute_features(G_struct)
features_struct["label"] = features_df["label"].values

X_s = features_struct[feature_cols].values
y_s = features_struct["label"].values

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
    X_s, y_s, test_size=0.3, random_state=42, stratify=y_s
)

clf_struct = RandomForestClassifier(n_estimators=200, random_state=42)
clf_struct.fit(X_train_s, y_train_s)
y_pred_s = clf_struct.predict(X_test_s)

print("=== After Structural Evasion Attack ===")
print(classification_report(y_test_s, y_pred_s))
```

```python
G_pois = G.copy()
max_node_id = max(G_pois.nodes())

num_new_bots = 50
new_bot_nodes = []

for i in range(1, num_new_bots + 1):
    nb = max_node_id + i
    new_bot_nodes.append(nb)
    G_pois.add_node(nb)

    neighbors = random.sample(human_nodes, 10)
    for h in neighbors:
        G_pois.add_edge(nb, h)


labels_dict_poison = labels.copy()
labels_dict_poison = list(labels_dict_poison)
labels_dict_poison.extend([1]*num_new_bots)


features_pois = compute_features(G_pois)
features_pois["label"] = labels_dict_poison

X_p = features_pois[feature_cols].values
y_p = features_pois["label"].values

X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(
    X_p, y_p, test_size=0.3, random_state=42, stratify=y_p
)

clf_pois = RandomForestClassifier(n_estimators=200, random_state=42)
clf_pois.fit(X_train_p, y_train_p)

y_pred_p = clf_pois.predict(X_test_p)

print("=== After Graph Poisoning Attack ===")
print(classification_report(y_test_p, y_pred_p))
```

```python
# train poisoned model again (already trained: clf_pois)

X_clean = features_df[feature_cols].values
y_clean = features_df["label"].values

y_clean_pred = clf_pois.predict(X_clean)

print("=== Poisoned model tested on CLEAN data ===")
print(classification_report(y_clean, y_clean_pred))
```
